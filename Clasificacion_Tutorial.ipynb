```python
import pandas as pd
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier

salary=pd.read_csv('/Programacion/Machine Learning/csv/ds_salaries.csv', sep=',')
salary.head()
salary['employment_type'].unique()
```

```python
salary.columns
x=salary.shape[0]
salary.dropna()
y=salary.shape[0]
print("Se eliminaron:",x-y," filas")

```

```python
#Creamos una copia del Dataset original
clean_data=salary.copy()
```

```python
clean_data['L_company_size_label']=(clean_data['company_size']=='L')*1
clean_data['M_company_size_label']=(clean_data['company_size']=='M')*0.5
clean_data['S_company_size_label']=(clean_data['company_size']=='S')*0
clean_data['EN_EL']=(clean_data['experience_level']=='EN')*0
clean_data['MI_EL']=(clean_data['experience_level']=='MI')*1
clean_data['SE_EL']=(clean_data['experience_level']=='SE')*2
clean_data['EX_EL']=(clean_data['experience_level']=='EX')*3
clean_data['High_Salary']=(clean_data['salary']>100000)*1
clean_data
```

```python
y=clean_data['High_Salary'].copy()
```

```python
features=['L_company_size_label',"M_company_size_label","S_company_size_label","EN_EL","MI_EL","SE_EL","EX_EL"]
x=clean_data[features].copy()
```

```python
X_train,X_test,y_train,y_test=train_test_split(x,y,test_size=0.33,random_state=324)
```

```python
salary_clasiffier=DecisionTreeClassifier(max_leaf_nodes=10,random_state=0)
```

```python
salary_clasiffier.fit(X_train,y_train)
```

```python
type(salary_clasiffier)
```

```python
predictions=salary_clasiffier.predict(X_test)
predictions[:20]
```

```python
X_test
```

```python
accuracy_score(y_true=y_test,y_pred=predictions)
```
